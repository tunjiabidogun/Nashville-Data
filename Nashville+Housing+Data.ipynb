```python
import pandas as pd
import numpy as np

df = pd.read_csv("Nashville_housing_data_2013_2016.csv.zip")
df.head()
```

```python
def getMonth(x):
    vals = x.split("-")
    return vals[1]

def getSeason(x):
    if x=="12":
        return "Winter"
    if x=="01":
        return "Winter"
    if x=="02":
        return "Winter"
    if x=="03":
        return "Spring"
    if x=="04":
        return "Spring"
    if x=="05":
        return "Spring"
    if x=="06":
        return "Summer"
    if x=="07":
        return "Summer"
    if x=="08":
        return "Summer"
    if x=="09":
        return "Fall"
    if x=="10":
        return "Fall"
    if x=="11":
        return "Fall"
```

```python
df['Month'] = df["Sale Date"].apply(getMonth)

df.head()
```

```python
df['Season'] = df["Month"].apply(getSeason)
df.tail()
```

```python
df_winter = df[df['Season'] == 'Winter']
    
df_winter['Sale Price'].mean()
```

```python
df_spring = df[df['Season'] == 'Spring']
df_summer = df[df['Season'] == 'Summer']
df_fall = df[df['Season'] == 'Fall']
```
